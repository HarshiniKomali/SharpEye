```python
import pandas as pd
```

```python
items = pd.read_csv('items.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
items.head()
```

```python
ratings.head()
```

```python
ratings = pd.merge(ratings,items,left_on='ProductID',right_on='product')
```

```python
ratings.head()
```

```python
ratings.describe()
```

```python
ratings.drop(['product_urls','search_links'],axis = 1, inplace=True)
```

```python
ratings.head()
```

```python
grouped_products = ratings.groupby('')
```

```python
grouped_products.groups.keys()
```

```python
for key in grouped_products.groups.keys():
  temp_df = grouped_products.get_group(key)
  result = temp_df.groupby('ProductID')['Rating'].mean()
  result_df = pd.DataFrame({'ProductID':result.index, 'Mean Rating':result.values})
  result_df = result_df.sort_values('Mean Rating',ascending=False)
  key_word_list = key.split("|")
  result_df.to_csv(key_word_list[-1]+'.csv',index=False)
```
